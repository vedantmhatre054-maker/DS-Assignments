```python
import pandas as pd

df = pd.read_csv("CSV files/grand_bazaar_data.csv")

print(df.head())
```

```python
# Q1. Find the Center of the Data
# Find Mean, Median and Mode of the dataset

print("Mean:", df["Order_Spend"].mean())
print("Median:", df["Order_Spend"].median())

print("Mode:")
print(df["Server_Hits_Per_Min"].mode())
```

```python
# Q2. How spread out is the Order Spend data?
# Find Variance and Standard Deviation

print("Variance:", df["Order_Spend"].var())

print("Standard Deviation:", df["Order_Spend"].std())
```

```python
# Q3. Is the data tilted or does it contain many outliers?
# Find Skewness and Kurtosis

print("Skewness:", df["Order_Spend"].skew())

print("Kurtosis:", df["Order_Spend"].kurt())
```

```python
# Q4. Import Visualization Libraries

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(
    data=df,
    x="Battery_Lifespan_Years",
    kde=True,
)

plt.title("Normal Distribution ")
plt.show()
```

```python
# Q4.2 Are Product Ratings equally distributed?
sns.histplot(
    data=df,
    x="Rating",
    bins=5
)
plt.title("Uniform Distribution")
plt.show()
```

```python
# Q4.3 Visualize Customer Spending Pattern
sns.histplot(
    data=df,
    x="Order_Spend",
    kde=True
)

plt.title("Exponential Distribution ")
plt.show()
```

```python
# Q4.4 Analyze Server Hits Per Minute

sns.histplot(
    data=df,
    x="Server_Hits_Per_Min",
    discrete=True
)

plt.title("Poisson Distribution")
plt.show()
```

```python
# Q5. Check if Order Spend is highly skewed

print(
    f"Original Order Spend Skewness: {df['Order_Spend'].skew():.2f}"
)
```

```python
# Q6. Apply Log Transformation

import numpy as np

df["Log_Order_Spend"] = np.log1p(df["Order_Spend"])

print(
    f"Transformed Order Spend Skewness: {df['Log_Order_Spend'].skew():.2f}"
)
```

```python
# Q7. Compare Before and After Transformation

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(
    data=df,
    x="Order_Spend",
    kde=True,
    ax=axes[0]
)

axes[0].set_title("Before Log Transformation")

sns.histplot(
    data=df,
    x="Log_Order_Spend",
    kde=True,
    ax=axes[1]
)

axes[1].set_title("After Log Transformation")

plt.show()
```
