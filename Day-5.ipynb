```python
#Q.2 Read both tables

import pandas as pd

df_cust = pd.read_csv("CSV files/customers.csv")
df_ord = pd.read_csv("CSV files/orders.csv")

print("Customer Table")
print(df_cust)

print("\n\nOrders Table")
print(df_ord)
```

```python
# Q.3 INNER JOIN

inner_merge = pd.merge(
    df_cust,
    df_ord,
    on= "customer_id",
    how = "inner",

)
print(inner_merge)
```

```python
#Q.4 Left Join keep right rows/ Right Join keep left rows

right_merge = pd.merge(
    df_cust,
    df_ord,
    on ="customer_id",
    how = "right",

)
print(right_merge)
```

```python
#Q.6 OUTER Join keeps everything from both tables
outer_merge = pd.merge(
    df_cust,
    df_ord,
    on = "customer_id",
    how = "outer",

)
print(outer_merge)
```

```python
#Q.7 Concat - Simply stics dataframe together    

import pandas as pd

batch_1 = pd.DataFrame({
    "customer_id": [10,11],
    "name": ["Vedant", "DANT"],
})

batch_2 = pd.DataFrame({
    "customer_id": [11,13,14],
    "name": ["OM" , "Ram", "Sam"],

})

combined_batch = pd.concat(
    [batch_1,batch_2],
    ignore_index=True  #without it duplicate index appears
    
)
print(combined_batch)

#Q.8 axis=1  is horizontal concat 
```

```python
#Q.9 
import pandas as pd

df_wide = pd.read_csv("CSV files/wide_sales.csv")

print(df_wide)
```

```python
#Q.9 Melt - Convert wide format to long format
df_long = pd.melt(
    df_wide,
    id_vars=["product"],#keep this column as it is
    value_vars=["Jan_Sales", "Feb_Sales", "Mar_Sales"],#convert this colums into rows
    var_name="Month",#Name the new column containing the sales values as Revenue.
    value_name="Revenue"#nam
)

print(df_long)
```

```python
#Q. 10  astype() Changing Data Types

import pandas as pd

df1 = pd.DataFrame({
    "id": [1, 2, 3],
    "Val": ["A", "B", "C"]
})

df2 = pd.DataFrame({
    "id": ["1", "2", "3"],
    "Price": [10, 20, 30]
})

print(df1)
print("\n")
print(df2)
```

```python
print(df1.dtypes)

print()

print(df2.dtypes)
```

```python
df2["id"] = df2["id"].astype(int)

print(df2.dtypes)
```

```python
# astype(int) Convert the id column from string (object) to integer (int64).

merged = pd.merge(
    df1,
    df2,
    on="id",
    how="inner"
)

print(merged)
```

```python
# Q. 12 left_on and right_on 

import pandas as pd

df_left = pd.DataFrame({
    "Client_ID": [101, 102],
    "Name": ["Amit", "Priya"]
})

df_right = pd.DataFrame({
    "Order_No": [5001, 5002],
    "User_ID": [101, 102]
})

result = pd.merge(
    df_left,
    df_right,
    left_on="Client_ID",
    right_on="User_ID",
    how="inner"
)

print(result)

#Use on= when both tables have the same column name.
#Use left_on= and right_on= when column names are different.
#Here Client_ID and User_ID contain the same values, so we tell Pandas how to match them
```

```python
#fillna() 

import pandas as pd

data = {
    "Name": ["Raj", "Priya", "Amit"],
    "Salary": [50000, None, 70000]
}

df = pd.DataFrame(data)

print("Before:")
print(df)

df["Salary"] = df["Salary"].fillna(0)

print("\nAfter:")
print(df)

#NaN means missing data.
#fillna(0) replaces all missing values with 0.
#Very useful when doing calculations because NaN can cause problems.
```

```python
#After you run this, create a new cell and run:

import pandas as pd

retail_cust = pd.DataFrame({
    "cid": [1, 2, 3],
    "Name": ["Raj", "Simran", "Karan"]
})

retail_orders = pd.DataFrame({
    "oid": [99, 98],
    "customer_link": [1, 2],
    "spend": [4500, 3200]
})

merged_retail = pd.merge(
    retail_cust,
    retail_orders,
    left_on="cid",
    right_on="customer_link",
    how="left"
)

merged_retail["spend"] = merged_retail["spend"].fillna(0)

print(merged_retail)
```
