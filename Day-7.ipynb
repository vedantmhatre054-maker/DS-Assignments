```python
# Step 2. Import Required Libraries

import pandas as pd
import scipy.stats as stats
```

```python
# Step 3. Load Dataset

df = pd.read_csv("CSV files/ab_testing_data.csv")

print(df.head())

print(df.columns)
```

```python
# Q1. Prepare Sample Data

spend_data = df["Spend_Amount"]
sample_mean = spend_data.mean()
sample_size = len(spend_data)

print("Sample Mean :", sample_mean)
print("Sample Size :", sample_size)
```

```python
standard_error = stats.sem(spend_data)
```

```python
# 3. Compute the 95% Confidence Interval

lower_bound, upper_bound = stats.t.interval(
    confidence=0.95,
    df=sample_size - 1,
    loc=sample_mean,
    scale=standard_error
)

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

```python
# Q2. Compare Region A and Region B

region_a = df[df["Region"] == "Region_A"]["Spend_Amount"]
region_b = df[df["Region"] == "Region_B"]["Spend_Amount"]

t_stat, p_val = stats.ttest_ind(
    region_a,
    region_b,
    equal_var=False
)

print("T Statistic:", t_stat)
print("P Value:", p_val)
```

```python
# Q3. Decide whether New Layout is better

if p_val < 0.05:
    print("Reject Null Hypothesis (H0)")
    print("New Layout has a significant effect.")
else:
    print("Fail to Reject Null Hypothesis (H0)")
    print("No significant difference found.")
```
