```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("logistics_data.csv")

print(df)
```

```python
matrix = df[["Jan_Tonnage", "Feb_Tonnage"]].to_numpy()

print(matrix)
```

```python
combined_weight = matrix[:, 0] + matrix[:, 1]

print("Combined Weight:")
print(combined_weight)
```

```python
processing_charge = matrix * 2

print("Processing Charge:")
print(processing_charge)
```

```python
value = matrix[2, 1]

print("Value at Row 2, Column 1:")
print(value)
```

```python
df["Freight_Cost"] = df["Freight_Cost"].replace("missing", 0)

print(df)
```

```python
df["Freight_Cost"] = df["Freight_Cost"].fillna(0)

print(df)
```

```python
df["Freight_Cost"] = df["Freight_Cost"].astype(int)

print(df.dtypes)
```

```python
mean_value = df["Freight_Cost"].mean()
median_value = df["Freight_Cost"].median()

print("Mean :", mean_value)
print("Median :", median_value)
```
