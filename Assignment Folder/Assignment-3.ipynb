## Task 2.1 - Hypothesis

**H₀ (Null Hypothesis):**
There is no significant difference in Spend_Amount between users who saw the Old_Layout and the New_Layout.

**H₁ (Alternative Hypothesis):**
There is a significant difference in Spend_Amount between users who saw the Old_Layout and the New_Layout.

```python
import sqlite3

conn = sqlite3.connect("bazaar_assignment.db")

df = pd.read_csv("../CSV files/bazaar_master_data.csv")

df.to_sql("bazaar_ledger", conn, if_exists="replace", index=False)

print("Database Created Successfully!")

conn.close()
```

```python
conn = sqlite3.connect("bazaar_assignment.db")

query = """
SELECT Customer_ID,
       Name,
       Spend_Amount
FROM bazaar_ledger
WHERE Spend_Amount >= 4000;
"""

result = pd.read_sql_query(query, conn)

print(result)

conn.close()
```

```python
conn = sqlite3.connect("bazaar_assignment.db")

query = """
SELECT Region,
       SUM(Spend_Amount) AS Total_Revenue
FROM bazaar_ledger
GROUP BY Region;
"""

result = pd.read_sql_query(query, conn)

print(result)

conn.close()
```
