```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "Traffic": [1200, 1450, 1100, 1300, 1700, 2200, 1900],
    "Revenue": [12000, 15500, 10500, 14000, 19500, 26000, 21000],
    "Platform_Tier": ["Basic", "Basic", "Standard", "Standard", "Premium", "Premium", "Premium"]
}

df = pd.DataFrame(data)

print(df)
```

```python
fig, ax = plt.subplots(figsize=(8,5))

ax.plot(
    df["Day"],
    df["Revenue"],
    color="blue",
    marker="^",
    linewidth=1
)

ax.set_title("7-Day Revenue Trend")
ax.set_xlabel("Day")
ax.set_ylabel("Revenue")

ax.grid(True)

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.barplot(
    data=df,
    x="Platform_Tier",
    y="Revenue",
    palette="deep"
)

plt.title("Average Revenue by Platform Tier")
plt.xlabel("Platform Tier")
plt.ylabel("Revenue")

plt.show()
```

```python
fig = px.scatter(
    df,
    x="Traffic",
    y="Revenue",
    color="Platform_Tier",
    hover_data=["Day"],
    title="Traffic vs Revenue"
)

fig.show()
```

```python
fig = go.Figure()

fig.add_trace(
    go.Bar(
        x=df["Day"],
        y=df["Traffic"],
        marker_color="green",
        name="Traffic"
    )
)

fig.update_layout(
    title="Daily Traffic Trend",
    template="plotly_white",
    xaxis_title="Day",
    yaxis_title="Traffic",
    xaxis_showgrid=True,
    yaxis_showgrid=True
)

fig.show()
```
