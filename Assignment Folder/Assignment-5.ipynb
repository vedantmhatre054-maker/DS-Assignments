```python
import pandas as pd
```

```python
snapshot_data = {
    "User_ID": ["USR_01", "USR_02", "USR_01", "USR_03", "USR_02", "USR_01", "USR_03"],

    "Date": pd.to_datetime([
        "2026-01-15",
        "2026-01-20",
        "2026-02-14",
        "2026-02-18",
        "2026-03-10",
        "2026-03-22",
        "2026-04-05"
    ]),

    "Revenue": [12000, 15500, 10500, 14000, 19500, 26000, 21000],

    "Platform_Tier": [
        "Basic",
        "Basic",
        "Standard",
        "Standard",
        "Premium",
        "Premium",
        "Premium"
    ]
}

df_extended = pd.DataFrame(snapshot_data)

print(df_extended)
```

```python
# Make Date the DataFrame index
df_monthly = df_extended.set_index("Date")

# Group data by month and calculate total revenue
monthly_revenue = df_monthly.resample("M")["Revenue"].sum()

# Display monthly revenue
print("Monthly Revenue")
print(monthly_revenue)
```

```python
# Create Order Month
df_extended["Order_Month"] = df_extended["Date"].dt.to_period("M")

# Create Cohort Month (first purchase month)
df_extended["Cohort_Month"] = (
    df_extended
    .groupby("User_ID")["Order_Month"]
    .transform("min")
)

# Calculate Months Elapsed
df_extended["Months_Elapsed"] = (
    df_extended["Order_Month"] -
    df_extended["Cohort_Month"]
).apply(lambda x: x.n)

# Display updated DataFrame
print(df_extended)

# Cohort Table
cohort_table = (
    df_extended
    .groupby(["Cohort_Month", "Months_Elapsed"])["User_ID"]
    .nunique()
    .reset_index(name="Active_Users")
)

print("\nCohort Analysis")
print(cohort_table)
```
