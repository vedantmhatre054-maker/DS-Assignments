```python
import pandas as pd
import numpy as np
from ydata_profiling import ProfileReport
```

```python
df_hr = pd.read_csv("CSV files/hr_techemployee_data.csv")
```

```python
df_hr.info()
```

```python
print(df_hr.columns)
```

```python
# Q2. Find Attrition Percentage

attrition_percentage = df_hr["Left_Company"].value_counts(normalize=True) * 100

print(attrition_percentage)
```

```python

# Q3. Salary Distribution Summary
print(df_hr["Monthly_Salary"].describe())
```

```python
# Q4. Compare Salary & Satisfaction based on Attrition

bivariate_summary = df_hr.groupby("Left_Company")[["Satisfaction_Score", "Monthly_Salary"]].mean()

print(bivariate_summary)
```

```python
# Q5. Department-wise Salary & Satisfaction Analysis

multivariate_pivot = df_hr.pivot_table(
    values=["Monthly_Salary", "Satisfaction_Score"],
    index=["Department"],
    columns=["Left_Company"],
    aggfunc=np.mean
)

print(multivariate_pivot)
```

```python
# Q6. Generate Automated EDA Report

automated_report = ProfileReport(
    df_hr,
    title="Grand Bazaar Automated HR Diagnostic",
    explorative=True
)
```

```python
automated_report.to_file("hr_automated_report.html")
```
